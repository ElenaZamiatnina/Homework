```python
with open('theDon.txt', encoding='windows-1251') as file:

    contents = file.read().lower()

def f(char):
    if char in [',', '.', '?', '!', '–', ';', ':', '*']:
        return ''
    else:
        return char

list(contents)
contents = ''.join(list(map(f, list(contents))))     

type(contents)
contents[0:30]
#len(contents)
contents = contents.split('\n')
type(contents)
contents[0:10]
contents = [i for i in contents if len(i)>0]
contents[0:10]

words = {}

for i in contents:
    for word in i.split():
        if word in words: 
            words[word] +=1
        else:
            words[word] = 1
len(words)
sorted_by_value = sorted(words.items(), key=lambda kv: -kv[1])
type(sorted_by_value)
sorted_by_value[:500]
for x in sorted_by_value[:500]:
    print(x[0])
```
